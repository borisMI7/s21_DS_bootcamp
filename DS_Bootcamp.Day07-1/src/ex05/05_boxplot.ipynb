# Boxplot

```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../../data/ab-test.csv")
```

```python
final_df = df.pivot(index = "uid", columns=["time", "group"], values="diff")
```

```python
after_list = [
    final_df[('after', 'control')].dropna(),
    final_df[('after', 'test')].dropna()
]
before_list = [
    final_df[('before', 'control')].dropna(),
    final_df[('before', 'test')].dropna()
]
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (15, 8), sharey=True)
fig.suptitle("Lag before labs deadlines. After and before the Newsfeed",
             fontsize=15)
ax1.set_title("after")
ax2.set_title("before")
ax1.set_xlabel("[group]")
ax2.set_xlabel("[group]")
ax1.grid(True)
ax2.grid(True)
ax1.boxplot(after_list, showmeans=True, tick_labels=["control", "test"], 
            boxprops = {'linewidth':3, 'color':'green'},
            medianprops = {'linewidth': 2, 'color': 'red'})
ax2.boxplot(before_list, showmeans=True, tick_labels=["control", "test"],
            boxprops = {'linewidth':3, 'color':'green'},
            medianprops = {'linewidth': 2, 'color': 'red'});
```
